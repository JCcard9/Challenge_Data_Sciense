### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python
faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

print(f'Faturamento Loja 1: R$ {faturamento_loja1:,.2f}')
print(f'Faturamento Loja 2: R$ {faturamento_loja2:,.2f}')
print(f'Faturamento Loja 3: R$ {faturamento_loja3:,.2f}')
print(f'Faturamento Loja 4: R$ {faturamento_loja4:,.2f}')
```

```python
import matplotlib.pyplot as plt

lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
faturamentos = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]
cores_faturamento = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99']

plt.figure(figsize=(10, 6))
plt.bar(lojas, faturamentos, color=cores_faturamento)
plt.title('Faturamento Total por Loja')
plt.ylabel('Faturamento (R$)')
plt.show()
```

# 2. Vendas por Categoria

```python
vendas_cat_loja1 = loja.groupby('Categoria do Produto')['Produto'].count().sort_values(ascending=False)
vendas_cat_loja2 = loja2.groupby('Categoria do Produto')['Produto'].count().sort_values(ascending=False)
vendas_cat_loja3 = loja3.groupby('Categoria do Produto')['Produto'].count().sort_values(ascending=False)
vendas_cat_loja4 = loja4.groupby('Categoria do Produto')['Produto'].count().sort_values(ascending=False)

print('--- Vendas por Categoria: Loja 1 ---')
print(vendas_cat_loja1)
print('\n--- Vendas por Categoria: Loja 2 ---')
print(vendas_cat_loja2)
print('\n--- Vendas por Categoria: Loja 3 ---')
print(vendas_cat_loja3)
print('\n--- Vendas por Categoria: Loja 4 ---')
print(vendas_cat_loja4)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
plt.pie(vendas_cat_loja1.values, labels=vendas_cat_loja1.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('hls', 8))
plt.title('Participação das Categorias nas Vendas - Loja 1')
plt.axis('equal')
plt.show()
```

# 3. Média de Avaliação das Lojas


```python
media_loja1 = loja['Avaliação da compra'].mean()
media_loja2 = loja2['Avaliação da compra'].mean()
media_loja3 = loja3['Avaliação da compra'].mean()
media_loja4 = loja4['Avaliação da compra'].mean()

print(f'Média Loja 1: {media_loja1:.2f}')
print(f'Média Loja 2: {media_loja2:.2f}')
print(f'Média Loja 3: {media_loja3:.2f}')
print(f'Média Loja 4: {media_loja4:.2f}')
```

```python
import matplotlib.pyplot as plt

lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
medias = [media_loja1, media_loja2, media_loja3, media_loja4]
cores = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99']

plt.figure(figsize=(10, 5))
plt.barh(lojas, medias, color=cores)

plt.title('Média de Avaliação por Loja')
plt.xlabel('Avaliação Média')
plt.xlim(0, 5)
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.gca().invert_yaxis()
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
def get_top_bottom(df, store_name):
    counts = df['Produto'].value_counts()
    top = counts.head(5)
    bottom = counts.tail(5)
    return top, bottom

top1, bot1 = get_top_bottom(loja, 'Loja 1')
top2, bot2 = get_top_bottom(loja2, 'Loja 2')
top3, bot3 = get_top_bottom(loja3, 'Loja 3')
top4, bot4 = get_top_bottom(loja4, 'Loja 4')

print('Análise de Produtos Finalizada.')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(18, 12))
fig.suptitle('Top 5 Produtos Mais Vendidos por Loja', fontsize=20)

sns.barplot(x=top1.values, y=top1.index, ax=axes[0,0], palette='viridis').set_title('Loja 1')
sns.barplot(x=top2.values, y=top2.index, ax=axes[0,1], palette='viridis').set_title('Loja 2')
sns.barplot(x=top3.values, y=top3.index, ax=axes[1,0], palette='viridis').set_title('Loja 3')
sns.barplot(x=top4.values, y=top4.index, ax=axes[1,1], palette='viridis').set_title('Loja 4')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(18, 12))
fig.suptitle('Top 5 Produtos Menos Vendidos por Loja', fontsize=20)

sns.barplot(x=bot1.values, y=bot1.index, ax=axes[0,0], palette='magma').set_title('Loja 1')
sns.barplot(x=bot2.values, y=bot2.index, ax=axes[0,1], palette='magma').set_title('Loja 2')
sns.barplot(x=bot3.values, y=bot3.index, ax=axes[1,0], palette='magma').set_title('Loja 3')
sns.barplot(x=bot4.values, y=bot4.index, ax=axes[1,1], palette='magma').set_title('Loja 4')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

# 5. Frete Médio por Loja

```python
frete_medio_loja1 = loja['Frete'].mean()
frete_medio_loja2 = loja2['Frete'].mean()
frete_medio_loja3 = loja3['Frete'].mean()
frete_medio_loja4 = loja4['Frete'].mean()

print(f'Frete Médio Loja 1: R$ {frete_medio_loja1:.2f}')
print(f'Frete Médio Loja 2: R$ {frete_medio_loja2:.2f}')
print(f'Frete Médio Loja 3: R$ {frete_medio_loja3:.2f}')
print(f'Frete Médio Loja 4: R$ {frete_medio_loja4:.2f}')
```

```python
import matplotlib.pyplot as plt

lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
fretes_medios = [frete_medio_loja1, frete_medio_loja2, frete_medio_loja3, frete_medio_loja4]
cores = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99']

plt.figure(figsize=(10, 6))
plt.bar(lojas, fretes_medios, color=cores)

plt.title('Custo Médio de Frete por Loja')
plt.ylabel('Frete Médio (R$)')
plt.xlabel('Loja')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

# Relatório de Análise e Recomendação

### 1. Introdução
O objetivo deste relatório é avaliar o desempenho das quatro lojas da rede para identificar oportunidades de melhoria e direcionar os esforços de vendas do Senhor João. A análise baseia-se em métricas de faturamento, satisfação do cliente, volume de vendas por categoria e custos logísticos.

### 2. Desenvolvimento e Análise dos Dados
Após o processamento dos dados das quatro unidades, observamos os seguintes pontos:

*   Faturamento: A Loja 1 apresenta o melhor desempenho financeiro (1.534.509,12) seguida pela loja 2 e 3. A Loja 4 possui o menor faturamento (1.384.497,58), indicando uma lacuna de aproximadamente  150 mil em relação à Loja 01.
*   Avaliação Média: Há uma consistência notável na satisfação dos clientes. Todas as lojas mantêm médias próximas a 4.0, com a Loja 3 ligeiramente à frente (4.05) e a Loja 1 com a menor média (3.98).
*   Categorias de Produtos: As categorias de Móveis e Eletrônicos são os pilares de vendas em todas as unidades. Produtos como 'Micro-ondas' e 'Cama box' aparecem consistentemente no Top 5 de vendas.
*   Frete Médio: A Loja 4 possui o custo de frete mais baixo  (31,28), enquanto a Loja 1 apresenta o maior custo logístico  (34,69).  

### 3. Conclusão e Recomendações
Com base na análise, sugerimos duas frentes de ação:

1. A loja que deve ser vendida é a Loja 4, devido ser a loja com menor faturamento.
2.  Foco de Expansão na Loja 4: Apesar de ter o menor faturamento, a Loja 4 possui o frete mais barato. Isso sugere uma eficiência logística que pode ser alavancada com investimentos em marketing e promoções para aumentar o volume de vendas, aproveitando a margem melhorada pelo baixo custo de envio.
3.  Manutenção da Liderança na Loja 1: Sendo a maior em faturamento, é vital investigar por que sua avaliação média é a mais baixa do grupo (3.98). Melhorar a experiência do cliente nesta unidade pode consolidar ainda mais sua posição de mercado.
